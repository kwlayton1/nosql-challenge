### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
#'scores': {'ConfidenceInManagement': 0, 'Hygiene': 0, 'Structural': 0}}
query = {'scores.Hygiene': 20}
  
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)

print("---------------------------------------------------------")
print("# of establishments with a hygiene score of 20 is", count)
print("---------------------------------------------------------")
print("Here is the first document:\n")

# retrieve establishments with Hygiene score of 20
hygiene_20 = list(establishments.find(query))

# Pretty print the first document
pprint(hygiene_20[0])
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Coordinates of "Penang Flavours"
latitude = 51.490142
longitude = 0.08384

# Search radius in degrees
degree_search = 0.01

# Create the query to find establishments with RatingValue of 5
query = {
    "RatingValue": 5,
    "geocode.latitude": {
        "$gte": latitude - degree_search,
        "$lte": latitude + degree_search
    },
    "geocode.longitude": {
        "$gte": longitude - degree_search,
        "$lte": longitude + degree_search
    }
}

# Sort by hygiene score
sort = [("scores.Hygiene", pymongo.ASCENDING)]

# retrieve establishments with Hygiene score of 20
rv5nearPF = list(establishments.find(query))

# Execute the query and limit the results to the top 5
results = establishments.find(query).sort(sort).limit(5)

# Print the results
for result in results:
    print("BusinessName:", result["BusinessName"])
    print("Hygiene Score:", result["scores"]["Hygiene"])
    print("Distance:", result["Distance"])
    print("----")
```

```python
# Convert result to Pandas DataFrame
rv5nearPF_df = pd.DataFrame(rv5nearPF)

# Display the number of rows in the DataFrame
num_rows2 = rv5nearPF_df.shape[0]
print("Number of rows in the DataFrame:", num_rows2)
print("Here are the first 10 rows of the DataFrame:\n")

# Display the first 10 rows of the DataFrame
rv5nearPF_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Define the aggregation pipeline
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": { "$sum": 1 }
        }
    },
    {
        "$sort": {
            "count": -1
        }
    }
]

# Execute the aggregation pipeline
hygiene0 = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(len(hygiene0))

# Print the first 10 results
for result in hygiene0[:10]:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
hygiene0_df = pd.DataFrame(hygiene0)

# Display the number of rows in the DataFrame
num_rows4 = hygiene0_df.shape[0]

print("Number of rows in the DataFrame:", num_rows4)
print("Here are the first 10 rows of the DataFrame:\n")

# Display the first 10 rows of the DataFrame
hygiene0_df.head(10)
```
